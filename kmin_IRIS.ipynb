```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score

# Memuat dataset IRIS
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Membuat model KMeans dengan 3 cluster (sesuai dengan jumlah kelas pada IRIS)
kmeans = KMeans(n_clusters=3)

# Melatih model KMeans
kmeans.fit(X)

# Mendapatkan label prediksi untuk setiap data point
labels = kmeans.labels_

# Membuat plot untuk visualisasi hasil clustering
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot sebenarnya
ax1.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
ax1.set_title('IRIS Dataset Ground Truth')

# Plot hasil clustering dengan KMeans
ax2.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
ax2.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='red')
ax2.set_title('KMeans Clustering')

plt.show()

# Menghitung Adjusted Rand Index (ARI)
ari = adjusted_rand_score(y, labels)
print("Adjusted Rand Index (ARI):", ari)

# Menghitung Adjusted Mutual Information (AMI)
ami = adjusted_mutual_info_score(y, labels)
print("Adjusted Mutual Information (AMI):", ami)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load data Iris
iris = load_iris()
data = iris.data

# Inisialisasi dan fitting model DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
predicted_labels = dbscan.fit_predict(data)

# Plot hasil clustering
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis', edgecolor='k', s=50)
plt.title('DBSCAN Clustering of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score

# Ground truth labels untuk data Iris
true_labels = iris.target

# Evaluasi dengan Adjusted Rand Index (ARI)
ari = adjusted_rand_score(true_labels, predicted_labels)
print("Adjusted Rand Index (ARI):", ari)
```
